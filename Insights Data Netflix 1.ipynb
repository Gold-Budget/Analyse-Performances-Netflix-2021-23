```python
flop_10 = dfdirt.groupby(["week", "show_title"])["weekly_hours_viewed"].sum().sort_values(ascending=False).tail(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(flop_10)
plt.title("Flop 10 Netflix")
plt.xlabel("Films les moins regardés")
plt.ylabel("Nombre de vues")
plt.show()
```

```python
flop_10 = flop_10.reset_index()
```

```python
plt.figure(figsize=(10,6))

plt.barh(flop_10["show_title"], flop_10["weekly_hours_viewed"])

plt.title("Flop 10 Netflix (Weekly Hours Viewed)")
plt.xlabel("Weekly Hours Viewed")
plt.ylabel("Show Title")

plt.show()
```

```python
dfdirt = pd.read_csv("netflix_top10.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr = dfdirt.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, 
cmap='coolwarm', fmt=".2f",
linewidths=0.5)
plt.title('Matrice de corrélation des données Netflix')
plt.savefig('Matrice de corrélation des données Netflix', dpi=300, bbox_inches='tight')
plt.show()
```

```python
flop_10 = dfdirt.groupby(["week", "show_title"])["weekly_hours_viewed"].sum().sort_values(ascending=False).tail(10)
flop_10 = flop_10.reset_index()
```

```python
plt.figure(figsize=(10,6))

plt.barh(flop_10["show_title"], flop_10["weekly_hours_viewed"])

plt.title("Flop 10 Netflix (Weekly Hours Viewed)")
plt.xlabel("Weekly Hours Viewed")
plt.ylabel("Show Title")

plt.savefig("Flop 10 Netflix (Weekly Hours Viewed)", dpi=300, bbox_inches='tight')
plt.show()
```

```python
top_10 = dfdirt.groupby(["week", "show_title"])["weekly_hours_viewed"].sum().sort_values(ascending=False).head(10)
top_10 = top_10.reset_index()
```

```python
vrai_top_10 = dfdirt.groupby("show_title")["weekly_hours_viewed"] \
  .sum() \
  .sort_values(ascending=False) \
  .head(10)
```

```python
vrai_top_10 = vrai_top_10.reset_index()
```

```python
plt.figure(figsize=(10,6))

plt.barh(vrai_top_10["show_title"], vrai_top_10["weekly_hours_viewed"])

plt.title("Top 10 Performances Netflix (Weekly Hours Viewed)")
plt.xlabel("Weekly Hours Viewed")
plt.ylabel("Show Title")

plt.savefig("Top 10 Performances Netflix (Weekly Hours Viewed)", dpi=300, bbox_inches='tight')
plt.show()
```
